```python
import pandas as pd
import numpy as np 

import pprint as pp
import re
```

```python
def chunks(l, n):
    """Yield successive n-sized chunks from l."""
    for i in range(0, len(l), n):
        yield l[i:i + n]
```

```python
f = open("ministerio de salud - tumores 2.csv", "r", encoding="utf8")
contents =f.read()

print("{:,}".format(len(contents)))

tum = contents.split(",")
print("{:,}".format(len(tum)))
```

```python
tum2 = []

for t in tum:
    if re.match("(\d*|-POB_DISTRITO)\s\d+", t):
        t2 = t.split(" ")
        tum2.extend(t2)
    else:
        tum2.append(t)
print(tum2[:100])
tum = tum2

    
```

```python
# Slices in 22 elements
tum_chunked = list(chunks(tum, 22))

print(tum_chunked[268:272])
```

```python
# Process df
columns_names = ["numero_caso", "codigo_subcapitulo", "subcapitulo", "codigo_dignostico", "diagnostico", 
                 "resultado_diagnostico", "sexo", "edad_quinquenal", "edad", "fecha_diagnostico", 
                 "provincia", "canton", "distrito", "region_ministerio_salud", "area_ministerio_salud", 
                 "region_ccss", "area_ccss", "codigo_establecimiento", "establecimiento", "poblacion_provincia", 
                 "poblacion_canton", "poblacion_distrito"]

tum_df = pd.DataFrame(tum_chunked[1:], columns=columns_names)
tum_df["fecha_diagnostico2"] = pd.to_datetime(tum_df["fecha_diagnostico"], unit="s", yearfirst=True)
tum_df["anno"] = tum_df["fecha_diagnostico2"].dt.year
tum_df = tum_df.replace({"":np.nan})
# tum_df = tum_df.dropna(axis=0, how="any")

print(tum_df.shape)

tum_df[tum_df.anno==2014].sample(10)
```

```python
# Write to csv

tum_df.to_csv("ministerio de salud - processed2.csv", sep=";", index=False)
# tum_df.to_excel("ministerio de salud - processed2.xlsx", index=False)
```

```python
# get frequency table
freq = pd.DataFrame(columns=["column", "value", "freq"])
freq_dfs = []
for column in tum_df.columns:
    
    freq = tum_df[column].value_counts().to_frame().reset_index()
    freq.columns = ["value", "counts"]
    freq["column"] = column
    freq["unique_values"] = len(freq)
    
#     display(freq)
    freq_dfs.append(freq)

    
freq = pd.concat(freq_dfs)
freq.sample(10)

freq.to_excel("frequency3.xlsx", index=False)
```

```python
# Replace comprised-like values to their lables
cols_to_replace = ['codigo_subcapitulo', 'subcapitulo', 'codigo_dignostico',
       'diagnostico', 'resultado_diagnostico', 'sexo', 'edad_quinquenal', 'provincia', 'canton', 'distrito',
       'region_ministerio_salud', 'area_ministerio_salud', 'region_ccss',
       'area_ccss', 'codigo_establecimiento', 'establecimiento']


dict_to_replace = {}
tum_to_replace = tum_df.dropna(axis=0, how="any").copy()

for column in cols_to_replace:
    
    techito = tum_to_replace[tum_to_replace[column].str.contains("\^")][column].drop_duplicates()
    techito_sort = techito.str.replace("\^", "").astype(int).sort_values()
    labels = tum_to_replace[~tum_to_replace[column].str.contains("\^")][column].drop_duplicates().sort_values()
    
    dict_column_values = {}

    for label, techo in zip(list(labels), list(techito_sort)):
        dict_column_values["^{}".format(techo)] = label

    dict_to_replace[column] = dict_column_values

    dict_column_values = {}
    
# pp.pprint(dict_to_replace)
```

```python
# Replace for each column in df
for column_name, values in dict_to_replace.items():
    tum_df[column_name] = tum_df[column_name].replace(to_replace=values)
    
tum_df.sample(10)
```

```python
# Drop irrelevant values
tum_df = tum_df.replace(to_replace=r"\^.*", value=np.nan, regex=True)
tum_df = tum_df.dropna(axis=0, how="any")
print(tum_df.shape)
```

```python
# Write to csv and excel 
tum_df.to_csv("ministerio de salud - processed3.csv", sep=";", index=False)
tum_df.to_excel("ministerio de salud - processed3.xlsx", index=False)
```
